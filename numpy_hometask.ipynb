```python
import numpy as np
import matplotlib.pyplot as plt
```

# Задание 1

```python
from itertools import cycle
from functools import partial

def f(x):
    return np.sin(x/5)*np.exp(x/10)+5*np.exp(-x/2)
def approx_f(x, w):
    return sum(x**i * wi for i, wi in enumerate(w))

points = [(1, 15), (1, 8, 15), (1, 4, 10, 15)]
res = [f]
for x in points:
    A = [[x_i**n for n in range(len(x))] for x_i in x]
    b = [f(x_i) for x_i in x]
    w = np.linalg.solve(A, b)
    f_t = partial(approx_f, w=w)
    res.append(f_t)

labels = cycle(['Исходная функция', "Многочлен первой степени", "Многочлен второй степени", "Многочлен третьей степени" ])
colors = cycle(['b', 'r', 'g', 'm'])
plt.figure(figsize=(15, 12))
xs = np.arange(0, 16, 16/200)
for f, clr, lbl in zip(res, colors, labels):
    ys = [f(x) for x in xs]
    plt.plot(xs, ys, color=clr, label=lbl,  linewidth=1)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

# Задание 2

```python
x = np.array(input().split(), dtype=np.number)
y = np.array(input().split(), dtype=np.number)

both_non_zero_objects = len(np.intersect1d(np.nonzero(x), np.nonzero(y)))
both_zero_objects = len(np.intersect1d(np.flatnonzero(x == 0), np.flatnonzero(y == 0)))

print('Answer:', both_non_zero_objects, len(x)-both_zero_objects)
```
